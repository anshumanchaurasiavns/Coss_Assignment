```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Associative.csv')
df
```

```python
cacheSize8 = df.iloc[0:3,]
cacheSize16 = df.iloc[3:7,]
cacheSize32 = df.iloc[7:12,]
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(cacheSize8['Block Size'], cacheSize8['Hit Ratio'], color='tab:blue',label="Cache Size=8")
ax.plot(cacheSize16['Block Size'], cacheSize16['Hit Ratio'], color='tab:orange',label="Cache Size=16")
ax.plot(cacheSize32['Block Size'], cacheSize32['Hit Ratio'], color='tab:green',label="Cache Size=32")
ax.legend()
plt.show()
```
