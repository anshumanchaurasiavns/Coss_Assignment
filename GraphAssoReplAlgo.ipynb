```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('AssoReplAlgo.csv')
df
```

```python
random = df.iloc[0:5,]
fifo = df.iloc[5:10,]
lru = df.iloc[10:15,]
random.info()
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(random['Cache Size'], random['Hit Ratio'], color='tab:blue',label="Random")
ax.plot(fifo['Cache Size'], fifo['Hit Ratio'], color='tab:orange',label="FIFO")
ax.plot(lru['Cache Size'], lru['Hit Ratio'], color='tab:green',label="LRU")
ax.legend()
plt.show()
```
